# Activation Functions

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Generate data for plotting
x = np.linspace(-10, 10, 400)
```

```python
def linear_activation(x):
        return x
x_values = range(-5, 5)
y_values = [linear_activation(x) for x in x_values]
plt.plot(x_values, y_values,label="linear_ac")
plt.xlabel('Input')
plt.ylabel('Output')
plt.title('Linear Activation Function')
plt.grid()
plt.legend()
plt.show()
```

```python
# Sigmoid function
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
plt.plot(x, sigmoid(x),label="sigmoid")
plt.title('Sigmoid Activation Function')
plt.xlabel('Input')
plt.ylabel('Sigmoid Output')
plt.grid()
plt.legend()
plt.show()
```

```python
# ReLU function
def relu(x):
    return np.maximum(0, x)
plt.plot(x, relu(x), color='orange',label="relu")
plt.title('ReLU Activation Function')
plt.xlabel("x")
plt.ylabel("ReLU(x)")
plt.legend()
plt.grid(True)
```

```python
# Tanh function
def tanh(x):
  return (np.exp(x) - np.exp(-x)) / (np.exp(x) + np.exp(-x))
plt.plot(x, tanh(x),  color='green')
plt.title('Tanh Activation Function')
plt.xlabel("x")
plt.ylabel("tanh(x)")
plt.grid(True)
```

```python
# Leaky ReLU function
def leaky_relu(x, alpha=0.01):
    return np.where(x > 0, x, alpha * x)

plt.plot(x, leaky_relu(x),  color='red')
plt.title('Leaky ReLU Activation Function')
plt.xlabel("x")
plt.ylabel("Leaky ReLU(x)")
plt.grid(True)
```

```python
def softmax(x):
    return np.exp(x) / np.sum(np.exp(x))
plt.plot(x, softmax(x),  color='red')
plt.title('Softmax Activation Function')
plt.xlabel("x")
plt.ylabel("softmax(x)")
plt.grid(True)
```
